```python
import pandas as pd 
```

```python
data = pd.read_csv("../gesture_classification_dataset.csv")
data.head()
```

```python
data.drop(columns= ['Unnamed: 0', 'confidence', 'id', 'hand', 'video'], inplace=True)
data.head()
```

```python
data.columns
```

```python
x = data.iloc[:, :-1]
y = data.iloc[:, -1]
```

```python
from sklearn.preprocessing import LabelEncoder
import pickle as pk

encoder = LabelEncoder()
y = encoder.fit_transform(y)

with open("label_encoder.pkl", "wb") as f:   
    pk.dump(encoder, f )
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
x_train.shape
```

```python
import torch

x_train_tensor = torch.tensor(x_train.values, dtype=torch.float32)
x_test_tensor = torch.tensor(x_test.values, dtype=torch.float32)

y_train_tensor = torch.tensor(y_train, dtype=torch.long)
y_test_tensor = torch.tensor(y_test, dtype=torch.long)
```

```python
y_train_tensor
```

```python
class GestureClassifier(torch.nn.Module):
    def __init__(self):
        super().__init__()

        self.network = torch.nn.Sequential(
            torch.nn.Linear(63, 64),
            torch.nn.ReLU(),

            torch.nn.Linear(64, 32),
            torch.nn.ReLU(),

            torch.nn.Linear(32, 5)
        )

    def forward(self, x):
        return self.network(x)
```

```python
model = GestureClassifier()

loss_fn = torch.nn.CrossEntropyLoss()

optimizer = torch.optim.Adam(model.parameters(), lr = 0.001)

epochs = 1000
```

```python
for t in range(epochs):
    
    optimizer.zero_grad()

    y_pred = model(x_train_tensor)

    loss = loss_fn(y_pred, y_train_tensor)

    if (t+1)%50 == 0:
        print(f"epoch: {t+1}, loss : {loss.item()}")

    loss.backward()

    optimizer.step()


    
```

```python
model.eval()

with torch.no_grad():
    train_outputs = model(x_train_tensor).argmax(dim=1).cpu().numpy()
    test_outputs =  model(x_test_tensor).argmax(dim=1).cpu().numpy()


train_outputs.shape
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print("Train accuracy:", accuracy_score(y_train, train_outputs))
print("Accuracy:", accuracy_score(y_test, test_outputs))
```

```python
print(confusion_matrix(test_outputs, y_test))
```

```python
print(classification_report(test_outputs, y_test))
```

```python
import json

folder = 'ver1/'

torch.save(model.state_dict(), folder+"state_dict.pth")

with open(folder+ "LabelEncoder.pkl" , "wb") as f:
    pk.dump(encoder, f)

config = {
    
    "num_features": 63,
    "num_classes": 5,
    "class_names": list(encoder.classes_)
}

with open(folder + "config.json", "w") as f:
    json.dump(config, f, indent=4)
```
